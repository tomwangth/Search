```python
### Constant Relative Risk Aversion Utility

import sympy as sp
import math 

e = math.e

Key = [] #Choice outcome/swtich point 

def CRRA(c, theta): #CRRA Function
    if theta > 0 and theta != 1:
        u1 = (1/(1-theta))*(c**(1-theta))
        return u1
    elif theta == 1:
        u2 = math.log(c,e)  
        return u2
    else:
        pass
    
    
def ExpectedUtility(p1, p2, c1, c2, theta1): #calculate expected utlity with CRRA utility 
    U1 = CRRA(c1,theta1)
    U2 = CRRA(c2, theta1)
    EU = p1*U1 + p2*U2
    return  EU


def OptionChoice(theta_1): #Given prob and value (expected value) in each Option, input theta and see
    for i in range(1,11):
        U_B = ExpectedUtility(i/10, 1-i/10, 3.85, 0.1, theta_1)
        U_A = ExpectedUtility(i/10, 1-i/10, 2, 1.6, theta_1)
        print (U_A, U_B)
        if U_A > U_B:
            Key.append('S')
        else:
            Key.append('R')
    return Key
    
test = input("Input Agent Coeffient") #Input agent coeffient to see
test_1 = float(test)

print(OptionChoice(test_1))

###graph of CRRA utility function (given different theta)

x_coords = []
y_coords = []

theta_value = input("Coefficient:")

for i in range (1,11):
    x_coords.append(i)
    theta_value1 = float(theta_value)
    y_coords.append(CRRA(i, theta_value1))
    

import pylab as pl

pl.plot(x_coords, y_coords, color = "blue", label = "Constant Relative Risk Aversion (CRRA) utility function")
pl.xlabel("c")
pl.ylabel("U(c)")
pl.legend()

pl.show()


    
```

```python
#  Search Model - Weitzman

import random
import math as m

class box(object):
    
    def __init__(self,reward_t,cost):
        self.prob_t = random.uniform(0.5,1)
        self.prob_f = 1-self.prob_t
        
        self.reward_t = reward_t
        self.reward_f = 0
        self.cost = cost

box_1 = box(random.uniform(30, 100), random.uniform(1, 10)) 
box_2 = box(random.uniform(30, 100), random.uniform(1, 10))
box_3 = box(random.uniform(30, 100), random.uniform(1, 10)) #generates 3 boxes

boxes = [box_1, box_2, box_3]
boxes_values = [box_1.reward_t, box_2.reward_t, box_3.reward_t] #list of box rewards values
boxes_costs = [box_1.cost, box_2.cost, box_3.cost]  #list of box costs



import sympy as sp


class Agent(object):

    def __init__(self, theta):
        self.theta = theta
    
    def reservation_price(self, p, x, c):
        
        z = sp.symbols("z")
        
        def CRRA(theta, value): #CRRA Utility
            if theta == 1.0:
                if value == 0.0:  # Handle value = 0.0
                    return 0.0
                else:
                    return m.log(value)
            elif theta > 0:
                if value == 0.0:  # Handle value = 0.0
                    return 0.0
                else:
                    return (1.0 / (1.0 - theta)) * (value) ** (1.0 - theta)
  
        f = (1 - p) * CRRA(self.theta, z - c) + p * CRRA(self.theta,x-c)- CRRA(self.theta, z)
        res_p = sp.solve(f, z) #solve for res_price
        return res_p
    
    
    def sort_lst_boxes(self):
        lst_res_p = []
        dict_res_p = {}
        j=0
        for box in boxes:
            lst_res_p.append(self.reservation_price(box.prob_t,box.reward_t,box.cost)[0])
            dict_res_p.update({j:self.reservation_price(box.prob_t,box.reward_t,box.cost)[0]})
            
            j=j+1
        
        s = sorted(dict_res_p, key=dict_res_p.get)
        
        lst_res_p.sort(reverse=True)
        return [s, lst_res_p]

    def open_boxes(self):
        reward_y = 0
        for box in boxes:
            get_reward = random.uniform(0,1)
            if box.prob_t<=get_reward<1:
                reward_y = box.reward_f
            elif 0<get_reward<box.prob_f:
                reward_y = box.reward_t
        return reward_y
   

    def search(self):
        lst_res_p = self.sort_lst_boxes()[1] #list of res price
        opened_box = [] 
        for i in range(0,3):
            max_res_p = lst_res_p[0]    
            
            if self.open_boxes() < max_res_p:
                lst_res_p.remove(lst_res_p[0]) #remove opened box each time
                reward_now = self.open_boxes()
                opened_box.append(self.sort_lst_boxes()[0][i]) #get a list of opened box
        
            else:
         
                print("Stop Search", "Reward = " + str(reward_now), "Opened Boxes =" + str(opened_box))
                pass
        
            
#test        
a1 = Agent(0.5)
print(a1.search())
```

##### Pig-Latin Rules

1. if the first letter of a word is a vowel (a, e, i, o, u), add the string 'way' to the end of the word, i.e., and becomes andway.
2. if the first letter of the word is a consanant move the string up to but not including the first vowel to the back of the word and add 'ay' to the end of the word, i.e., Kevin becomes evinKay.

```python
class Welcome(object):
    """ class desciptor """
    # class data types here
    def __init__(self, name):  # class initializer called when object created

        
        self.name = name   # the way other methods access name as well
        pig_latin_name = "" # initialize a pig_latin_name      change here
    
    def show(self): 
    
        print ("Hi {}!".format(self.name))
        
    def pig_latin(self):
        def pig_latinize(word):
            vowels = "aeiouAEIOU"
            if word == "":
                return word
            if word[0] in vowels:
                return word + 'way '
            else:
                for index, letter in enumerate(word):
                    if letter in vowels:
                        return word[index:] + word[:index] + "ay "
                    
        persons_name = self.name.split(" ")
        self.pig_latin_name = ""                        # change here
        for word in persons_name:
            self.pig_latin_name += pig_latinize(word)   # change here
        # pig_latin_name = "evinKay abeMcCay"
        print ("iHay {}".format(self.pig_latin_name))   # change here
 

welcome_carolyn = Welcome("Carolyn Ann Drislane") # make an object
welcome_carolyn.show()  # say hi      
welcome_carolyn.pig_latin() # say hi in pig latin
print()
welcome_kathy = Welcome("Kathleen Byrne McCabe") # make an object
welcome_kathy.show()  # say hi      
welcome_kathy.pig_latin() # say hi in pig latin
print()
welcome_kevin = Welcome("Kevin Alfred McCabe") # make an object
welcome_kevin.show()  # say hi      
welcome_kevin.pig_latin() # say hi in pig latin
print()
print(welcome_kevin.name)
print(welcome_kevin.pig_latin_name)
```
