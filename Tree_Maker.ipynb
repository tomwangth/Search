```python
import sympy as sp

class BoxSet():
    def __init__(self, box_list=[]):
        '''
            Input:
                box_list (list): list containing boxes in a set. If left empty, the list will just be an empty list

        '''
        self.box_list = box_list  # list to store all boxes added

    def __str__(self):
        '''format box set to print'''

        list_info = []
        for b in self.box_list:
            list_info.append(b.get_info())
        return str(list_info)

    def add_box(self, box_to_add):
        '''
            Input:
                box (list): a box to be added to the list

        '''

        self.box_list.append(box_to_add)  # append box to list
        return None

    def get_box(self):
        '''return the list of boxs'''

        return self.box_list


class Box():
    def __init__(self, name, prob, hip, lop, cost):
        '''
            Input:
                prob (int || float):  The probability of the box giving the high payout
                   ^ Note: This is short hand for probability.

                hip(int || float): The value of the high payoff.
                    ^This is shorthand for high_payoff.

                lop(int || float): The value of the low payoff.
                    ^This is shorthand for low_payoff.
                cost(int || float): The value of the cost to open.
        '''

        # checks to make sure all the inputs are numbers
        if ((isinstance(prob, int) or isinstance(prob, float)) and (isinstance(hip, int) or isinstance(hip, float)) \
                and (isinstance(lop, int) or isinstance(lop, float)) and (
                        isinstance(cost, int) or isinstance(cost, float))):
            self.prob = prob
            self.hip = hip
            self.lop = lop
            self.cost = cost
        else:
            raise ValueError("Please make sure all inputs are ints or floats")

    def __str__(self):
        '''format to print'''
        return "prob: %d, high_payoff: %d, low_payoff: %d, cost:%d" % (self.prob, self.hip, self.lop, self.cost)

    def get_prob(self):
        '''returns the probability of the box'''
        return self.prob

    def get_hip(self):
        '''returns the high payoff of the box'''
        return self.hip

    def get_lop(self):
        '''returns the low payoff of the box'''
        return self.lop

    def get_cost(self):
        '''returns the cost of the box'''
        return self.cost

    def get_info(self):
        dic = {}
        dic["probability"] = self.prob
        dic["high_payoff"] = self.hip
        dic["low_payoff"] = self.lop
        dic["cost"] = self.cost
        return dic
        # return "prob: %d, high_payoff: %d, low_payoff: %d, cost:%d"%(self.prob,self.hip,self.lop,self.cost)

box1 = Box(.5,100,0,20)
box2 = Box(.9,0,0,0)
box3 = Box(.4,100,30,10)

box_set = BoxSet([box1,box2,box3])

indi_box = box_set.get_box()
print(indi_box[0].get_cost())


class MakeTree_res_p:

    def __init__(self):

        self.prob = box.get_prob()
        self.hip = box.get_hip()
        self.lop = box.get_lop()
        self.cost =  box.get_cost()






    def res_p(prob, reward, cost):
        z = sp.symbols("z")
        f = (prob * reward - cost) / prob - z
        res_p = sp.solve(f, z)
        return res_p

    res_p_lst = []
    dict_res_p = {}
    j=0


    boxes = box_set.get_box()



    for box in boxes:
        z = res_p(box.get_prob(), box.get_hip(), box.get_cost())
        res_p_lst.append(z)
        dict_res_p.update({j: res_p(box.get_prob(), box.get_hip(), box.get_cost())})

        j = j + 1

    s = sorted(dict_res_p, key=dict_res_p.get)  #list of boxes number in order(from w/ highest z to lowest z)

    res_p_lst.sort(reverse = True) #list of reservation price


    def start():
        file_name = input('Enter a file name ')
        try:
            file = open(file_name + '.txt', 'r')
            file.close()
            print("File {} already exists".format(file_name))
            return  # if it does return
        except IOError:
            print('File {} will now be written.'.format(file_name))

        file = open(file_name + '.txt', 'w')
        return file

    def close_file(file):
        file.close()
        print("file has been saved")

    def node_output(file, node, name, type, descendants, probability, pay):
        """writes out the information for a node dictionary"""
        file.write('node,' + str(node) + '\n')
        file.write('name,' + name + '\n')
        file.write('ancestor,-1' + '\n')
        file.write('filled,True' + '\n')
        file.write('used,False' + '\n')
        file.write('subvale,0.0' + '\n')
        file.write('type,' + type + '\n')
        if type == 't':
            file.write('pay,' + str(pay) + '\n')
        else:
            des_length = len(descendants)
            file.write('length,' + str(des_length) + '\n')
            for j in range(des_length):
                file.write('descendant,' + str(descendants[j]) + '\n')
            if type == 'n':
                for j in range(des_length):
                    file.write('prob,' + str(probability[j]) + '\n')

    def node_save(node, name, type, descendants, probability, pay):
        """shows node information being written to file"""
        print('node, {}'.format(str(node)))
        print('name, {}'.format(name))
        print('ancestor, -1')
        print('filled, True')
        print('used, False')
        print('subvale, 0.0')
        print('type, {}'.format(type))
        if type == 't':
            print('pay, {}'.format(str(pay)))
        else:
            print('length, {}'.format(str(len(descendants))))
            for k in range(len(descendants)):
                print('descendant, {}'.format(str(descendants[k])))
            if type == 'n':
                print('prob, {}'.format(str(probability[0])))
                print('prob, {}'.format(str(probability[1])))
        print

    def check_state(name, state):
        if name[-1:] in ['W', 'L']:
            return state
        else:
            if name[-1:] in name[:len(name) - 1]:
                return 't'
            else:
                return state

    def build_payoff(name, payoffs, low_payoffs, costs):
        k = int(name[-1:])  # get index of box
        j = name.find(name[-1:])  # find where it first occurs in string
        if name[j + 1] == "W":
            pay = payoffs[k - 1]
        else:
            pay = low_payoffs[k - 1]
        # determine costs right now brute force for example
        # TODO fix this for costs lists
        cost = 0.0
        for k in range(len(costs)):
            if str(k + 1) in name:
                cost += costs[k]
        return pay - cost

    def get_probability(name, probabilities):
        k = int(name[-1:]) - 1
        return [probabilities[k], 1.0 - probabilities[k]]

    def get_descendants(type, dnodes):
        descendants = [dnodes, dnodes + 1]
        if type == 'd':
            descendants.append(dnodes + 2)
        return descendants

    def get_new_labels(old_labels, state):
        new_labels = []
        for label in old_labels:
            if check_state(label, state) == 't':
                continue
            if state == 'n':
                new_labels.append(label + 'W')
                new_labels.append(label + 'L')
            else:
                new_labels.append(label + str(s[0]))
                s.remove(s[0])

        return new_labels

    def parse_level(file, state, old_labels, next_node, probabilities, payoffs, low_payoffs, costs):
        """uses old_labels and state to determine new_labels then outputs node information to save  to file

            arg: file, file object for saving tree information
            arg: state, string, switches back and forth between 'n' and 'd'
            arg: old_labels, list, of all node names used so far
            arg: next_node, integer, next node to be saved
            arg: probabilities, list, probabilities of payoffs in boxes
            arg: payoffs, list, payoffs associcated with boxes
            arg: costs, list, cost to open boxes

            return: new_labels, list of old_labels updated"""

        dnodes = next_node + len(old_labels)
        for j, k in enumerate(old_labels):
            node = next_node + j
            name = k
            type = check_state(name, state)
            if type == 't':
                pay = build_payoff(name, payoffs, low_payoffs, costs)
            else:
                pay = 0.0
            descendants = [0, 0, 0]
            probability = [0, 0]
            if type == 'd' or type == 'n':
                descendants = get_descendants(type, dnodes)
                dnodes = dnodes + len(descendants)
            if type == 'n':
                probability = get_probability(name, probabilities)
            node_save(node, name, type, descendants, probability, pay)
            node_output(file, node, name, type, descendants, probability, pay)

        new_labels = get_new_labels(old_labels, state)
        return new_labels


    def create_dictionary(self):
        dic_tree = [{'name': 'start', 'type': 'd', 'descendants': [1], 'ancestor': -1}]
        dic_now = {}
        n = len(res_p_lst)
        if n >= 2:
            n_level = 5 + 2*(n-2)

        for i in range (1, n_level):
            if  i % 2 == 0:
                state = "d"
            elif i % 2 == 1:
                state = "n"
            level = i


        next_node = next_node + len(old_labels)
        old_labels = new_labels
        new_labels = level_starter(file, level, old_labels, state, next_node, probabilities, payoffs, low_payoffs, costs)


        
```
